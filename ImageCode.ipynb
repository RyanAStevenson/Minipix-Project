```python
import numpy as np
from PIL import Image
import pandas as pd
from matplotlib import pyplot as plt
from PIL import Image

directory = "D:\\MiniPix Data\\Energy Experiment\\6\\Outside\\"
# loops over all files in directory
for i in range(6000):
    if i < 10:
        file_name = "Outside00" + str(i)
    elif i > 9 and i < 100:
        file_name = "Outside0" + str(i)
    elif i > 99 and i < 1000:
        file_name = "Outside" + str(i)
    elif i > 999:
        file_name = "Outside" + str(i)
    file = open(directory + file_name + ".txt")

    # read in data
    data = pd.read_table(file, header=None, delimiter=' ', nrows=256)
    data = data.iloc[:, :]
    rows, cols = data.shape
    na = np.array(data)
    # remove dead pixels with value of 0.5
    na[na == 0.5] = 0
    total = np.sum(na)

    ## make an image of the canvas
    plt.figure(figsize=(3.4, 3.4), dpi=100) # dimensions give a 256x256 pixel image
    plt.axis('off')
    plt.imshow(na>0, origin="lower",cmap="gray")
    # saves image if there is a particle i.e. sum of pixels values > 0
    if total > 0:
        # save image
        plt.savefig("D:\\MiniPix Data\\Energy Experiment\\6\\Outside (Images)\\"
                    + file_name + ".jpg",dpi=100, bbox_inches='tight',pad_inches = 0)
        # save image as a grayscale image as as opposed to an rgb image (which is what the previous step does)
        Image.open("D:\\MiniPix Data\\Energy Experiment\\6\\Outside (Images)\\"
                    + file_name + ".jpg").convert("L").save("D:\\MiniPix Data\\Energy Experiment\\6\\Outside (Images)\\"
                    + file_name + ".jpg")
        # saves frame name to txt file
        txt_file = open("D:\\MiniPix Data\\Energy Experiment\\6\\Outside (Images)\\File Names.txt", "a")
        txt_file.write(file_name +" "+ str(total) + "\n")
        txt_file.close()

        # read in image data
        directory1 = "D:\\MiniPix Data\\Energy Experiment\\6\\Outside (Images)\\"
        img = plt.imread(directory1 + file_name + ".jpg")

        # checks that there is a particle i.e. sum of pixels values > 0
        total1 = 0
        for i in range(256):
            for j in range(256):
                total1 += img[i][j]
        # if a pixel has a non-zero value, append the x and y coordinate to separate arrays
        if total1 > 0:
            x_arr, y_arr = [], []
            for i in range(256):
                for j in range(256):
                    if img[i][j] > 40:
                        x_arr.append(j)
                        y_arr.append(i)
            # get the min and max x and y values to find the limits of the non-zero pixels within the canvas
            x_min, x_max = min(x_arr), max(x_arr)
            y_min, y_max = min(y_arr), max(y_arr)

            # create range in which the particle (and all non-zero particles) lie
            x_range, y_range = (x_max - x_min) + 1, (y_max - y_min) + 1
            #print(x_min, x_max, y_min, y_max, x_range, y_range)

            # recreate image in only the range provided by the particle and non-zero pixels instead of the whole canvas
            rows, cols = (x_range, y_range)
            pix_arr = [[0]*rows]*cols

            pix_arr = np.array(pix_arr)

            # get infromation from the whole canvas and add the pixel values to a new array
            for x in range(x_range):
                for y in range(y_range):
                    x_val = x_min + x # the positions of the particle will likely have some offset (i.e. wont start at 0,0) so the code has to go though the old positions with the offset but start at 0,0 when adding the code to a new array
                    y_val = y_min + y
                    pix_arr[y][x] = img[y_val][x_val]
                    
            # calculate the length and zenith angle        
            length = np.sqrt(((x_range*0.05078125)**2)+((y_range*0.05078125)**2))
            zenith_angle = np.arctan(y_range/x_range)
            if length > 0.40625: #minimum length of assumed muon. Here it is set to 8 pixels
                # plot and save the image
                plt.figure(figsize=(3.4, 3.4), dpi=100)
                plt.axis('off')
                plt.imshow(pix_arr,cmap="cividis")

                plt.savefig("D:\\MiniPix Data\\Energy Experiment\\6\\test\\"
                                    + file_name +" "+str(zenith_angle)+" "+ str(length) +" "+ str(total) +".jpg",dpi=100, bbox_inches='tight',pad_inches = 0)
                plt.close()
             # saves frame number, angle, length and energy to txt file
                txt_file = open("D:\\MiniPix Data\\Energy Experiment\\6\\test\\Muons.txt", "a")
                txt_file.write(file_name +" "+str(zenith_angle)+" "+ str(length) +" "+ str(total) + "\n")
                txt_file.close()
    plt.close()
```
